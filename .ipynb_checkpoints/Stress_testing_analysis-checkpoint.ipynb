```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import style
style.use("ggplot")
```

```python
df=pd.read_excel("Summary.xlsx")
```

```python
df
```

```python
df.dropna(inplace=True,axis=0)
```

```python
df
```

```python
df["avg_test_time"]=pd.to_numeric(df["avg_test_time"])
```

```python
user_id=df["Number_Users"].values.tolist()
img_1_min=df["img_1_min"].values.tolist()
img_1_max=df["img_1_max"].values.tolist()
y1=list(zip(img_1_min,img_1_max))


user_id=df["Number_Users"].values.tolist()
img_2_min=df["img_2_min"].values.tolist()
img_2_max=df["img_2_max"].values.tolist()
y2=list(zip(img_2_min,img_2_max))



user_id=df["Number_Users"].values.tolist()
img_3_min=df["img_1_min"].values.tolist()
img_3_max=df["img_1_max"].values.tolist()
y3=list(zip(img_3_min,img_3_max))


(fig,[ax1,ax2,ax3])=plt.subplots(1,3)
sns.scatterplot(data=df,x="Number_Users", y="img_1_avg",ax=ax1,color="red",marker="o")
sns.lineplot(data=df,x="Number_Users", y="img_1_avg",ax=ax1,color="red")
ax1.plot((user_id,user_id),([i for (i,j) in y1], [j for (i,j) in y1]),c='black',linestyle="dashdot",marker="_")
ax1.set_title("Image-1")

sns.scatterplot(data=df, x="Number_Users", y="img_2_avg",ax=ax2,color="blue",marker="o")
sns.lineplot(data=df, x="Number_Users", y="img_2_avg",ax=ax2,color="blue")
ax2.plot((user_id,user_id),([i for (i,j) in y2], [j for (i,j) in y2]),c='black',linestyle="dashdot",marker="_")
ax2.set_title("Image-2")

sns.scatterplot(data=df, x="Number_Users", y="img_3_avg",ax=ax3,color="green",marker="o")
sns.lineplot(data=df, x="Number_Users", y="img_3_avg",ax=ax3,color="green")
ax3.plot((user_id,user_id),([i for (i,j) in y3], [j for (i,j) in y3]),c='black',linestyle="dashdot",marker="_")
ax3.set_title("Image-3")

ax1.set_ylim(0,25)
ax2.set_ylim(0,25)
ax3.set_ylim(0,25)
fig.set_size_inches(18.5, 10.5)
```

```python
plt.show()
```

```python
with sns.axes_style("white"):
    sns.scatterplot(data=df, x="Number_Users", y="avg_test_time",color="black")
    sns.lineplot(data=df, x="Number_Users", y="avg_test_time",color="black",label="Avg_Test_Time")
    ax2 = plt.twinx()
    sns.scatterplot(data=df, x="Number_Users", y="CPU",color="red")
    sns.lineplot(data=df, x="Number_Users", y="CPU",color="red",label="CPU")
fig=plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
sns.lineplot(data=df, x="Number_Users", y="AVG server Response",color="black")
sns.scatterplot(data=df, x="Number_Users", y="AVG server Response",color="red")
fig=plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
sns.lineplot(data=df, x="Number_Users", y="Total Counts",color="black")
sns.scatterplot(data=df, x="Number_Users", y="Total Counts",color="red")
fig=plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
import numpy as np
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr, mask=mask, square=True,annot=True)
f.set_size_inches(18.5, 10.5)
```
